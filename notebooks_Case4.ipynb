```python
#нужно написать программу, которая принимает на вход файл с данными о пользователях и их друзьях в одной из социальных сетей,
#выводит на экран количество пользователей, у которых больше 100 друзей. Кроме того, программа должна строить круговую диаграмму,
#показывающую процентное соотношение пользователей по количеству друзей: менее 10, от 10 до 50, от 50 до 100 и более 100.
```

```python
#Подключение библиотек
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as mb
```

```python
#Чтение и вывод фрейма
network = pd.read_csv('Social.csv', delimiter=',')
print(network)
```

```python
#выведение на экран количества пользователей, у которых больше 100 друзей.
famous = len(network.loc[network['friends'] > 100, 'name'])
print('Количество пользователей с более чем 100 друзьями:',famous)
```

```python
#Вычисление процента градации: менее 10, от 10 до 50, от 50 до 100 и более 100
a = int(len(network.loc[network['friends'] < 10, 'name']) / (len(network['name'])) * 100)
b = int(len(network[(network["friends"] <= 50) & (network["friends"] >= 10)]) / (len(network['name'])) * 100)
c = int(len(network[(network["friends"] <= 100) & (network["friends"] >= 50)]) / (len(network['name'])) * 100)
d = int(len(network.loc[network['friends'] > 100, 'name']) / (len(network['name'])) * 100)
print(a, b,c,d)
```

```python
#Создание графика

#Создание списков
users = ['менее 10','от 10 до 50','от 50 до 100','более 100']
procent = [a,b,c,d]


# создание круговой диаграммы
mb.pie(procent, labels= users, autopct="%1.1f%%")
mb.title("Круговая диаграмма пользователей по количеству друзей:")
mb.show()
```
